```python
#single person image
import cv2
import matplotlib.pyplot as plt


img = cv2.imread(r"D:\me.jpg",0)

# Ensure the image is loaded correctly
if img is None:
    print("Error: Could not open or find the image.")
else:
    # Display the original image with its original colors
    plt.figure(figsize=(10, 10))
    plt.imshow(cv2.cvtColor(img, cv2.COLOR_BGR2RGB))  # Convert BGR to RGB for correct display
    plt.title('Original Image')
    plt.axis('off')  # Turn off axis labels
    plt.show()



    
# Canny edge detection   --- to identify edges in the images
edges = cv2.Canny(img, 100, 100)
plt.figure(figsize=(10,10))
plt.imshow(edges, cmap='gray')
plt.title('Canny Edge Detection')
plt.xticks([]), plt.yticks([])
```

```python
#circuit

import cv2
import numpy as np
from matplotlib import pyplot as plt


import numpy as np
from matplotlib import pyplot as plt

img = cv2.imread(r"D:\me.jpg", 0)



# Canny edge detection   --- to identify edges in the images
edges = cv2.Canny(img, 50, 70)
plt.subplot(2, 2, 1)
plt.imshow(edges, cmap='binary')
plt.title('Canny Edge Detection')
plt.xticks([]), plt.yticks([])



# Image resizing
resized = cv2.resize(img, (90,80))
plt.subplot(2,2,2)
plt.imshow(resized, cmap='gray')
plt.title('Resized Image')
plt.xticks([]), plt.yticks([])

# Image cropping
cropped = img[50:200, 50:200]
plt.subplot(2,2,3)
plt.imshow(cropped, cmap='gray')
plt.title('Cropped Image')
plt.xticks([]), plt.yticks([])

# Color space conversion (Grayscale to RGB)
rgb_image = cv2.cvtColor(img, cv2.COLOR_GRAY2RGB)
plt.subplot(2,2,4)
plt.imshow(rgb_image)
plt.title('Grayscale to RGB')
plt.xticks([]), plt.yticks([])

```

```python


# Blurring (Gaussian Blur)
blurred = cv2.GaussianBlur(img, (5, 5), 0)
plt.subplot(2,2,1)
plt.imshow(blurred, cmap='gray')
plt.title('Gaussian Blur')
plt.xticks([]), plt.yticks([])

# Thresholding (Simple Thresholding)
_, thresh = cv2.threshold(img, 127, 255, cv2.THRESH_BINARY)
plt.subplot(2,2,2)
plt.imshow(thresh, cmap='gray')
plt.title('Thresholding')
plt.xticks([]), plt.yticks([])

# Morphological Operations (Erosion)
kernel = np.ones((5, 5), np.uint8)
erosion = cv2.erode(img, kernel, iterations=1)
plt.subplot(2,2,3)
plt.imshow(erosion, cmap='gray')
plt.title('Erosion')
plt.xticks([]), plt.yticks([])

# Image Transformation (Rotation)
rows, cols = img.shape
matrix = cv2.getRotationMatrix2D((cols / 2, rows / 2), 50, 1)
rotated = cv2.warpAffine(img, matrix, (cols, rows))
plt.subplot(2,2,4)
plt.imshow(rotated, cmap='gray')
plt.title('Image Rotation')
plt.xticks([]), plt.yticks([])

plt.show()
```

```python
# Contour detection
contours, _ = cv2.findContours(edges, cv2.RETR_TREE, cv2.CHAIN_APPROX_SIMPLE)
contour_img = cv2.cvtColor(img, cv2.COLOR_GRAY2BGR)
cv2.drawContours(contour_img, contours, -1, (0, 255, 0), 2)

plt.figure(figsize=(10, 10))
plt.imshow(contour_img)
plt.title('Contours')
plt.xticks([]), plt.yticks([])
plt.show()



```

```python
# Face Detection (using Haar cascades)
face_cascade = cv2.CascadeClassifier(cv2.data.haarcascades + 'haarcascade_frontalface_default.xml')
faces = face_cascade.detectMultiScale(img, scaleFactor=1.1, minNeighbors=5)
face_img = cv2.cvtColor(img, cv2.COLOR_GRAY2BGR)
for (x, y, w, h) in faces:
    cv2.rectangle(face_img, (x, y), (x + w, y + h), (255, 0, 0), 2)

plt.figure(figsize=(10, 10))
plt.imshow(face_img)
plt.title('Face Detection')
plt.xticks([]), plt.yticks([])
plt.show()
```

```python
# Feature Detection (ORB)
orb = cv2.ORB_create()
keypoints, descriptors = orb.detectAndCompute(img, None)
feature_img = cv2.drawKeypoints(img, keypoints, None, color=(0, 255, 0))

plt.figure(figsize=(10, 10))
plt.imshow(feature_img)
plt.title('ORB Feature Detection')
plt.xticks([]), plt.yticks([])
plt.show()
```
